### 5.1 KLダイバージェンス
### 5.1.1 準備(数式の表記について)
1. 期待値
    - 連続型の確率変数$x$があり、その確率密度が$p(x)$.
    - このとき、関数$f(x)$の期待値は次式で表される
    $$
    \bm{E}_{p(x)}[f(x)] = \int f(x)p(x)dx  \quad \text{($p(x)$に関する期待値であることを明示)}
    $$
1. パラメータの表記場所
    $$
    p(x;\bm{\theta}) \rightarrow p_\theta(x)    \quad \text{(\textbf{$\bm{\theta}$} = \{ \textbf{$\phi$}, \textbf{$\mu$}, \textbf{$\Sigma$} \})}
    $$

--- 

### 5.1.2 KLダイバージェンスの定義式
#### 定義
2つの確率分布$p(x), q(x)$が与えられた時、KLダイバージェンスは次式で表される.
>$$
D_{KL} (\;p \;|| \;q\;) = \int p(x) \; \log\cfrac{p(x)}{q(x)} \;dx \quad (xが連続型の確率変数の場合)
$$
> $$
D_{KL} (\;p \;|| \;q\;) = \sum p(x) \; \log\cfrac{p(x)}{q(x)} \;dx \quad (xが離散型の確率変数の場合)
$$
#### 特性
1. 2つの確率分布が異なるほど大きな値を示す. ($\because$)
1. 0以上の値を取り、2つの確率分布が同じときのみ0になる. ($\;\because \frac{p(x)}{q(x)} = 1 $)
1. 非対称な尺度であるため、$D_{KL} (\;p \;|| \;q\;) \neq D_{KL} (\;q \;|| \;p\;)$

#### 具体例（コインの表と裏のでる確率）
1. ケースA

    |  | 表の出る確率 | 裏の出る確率 |
    |----|----|----|
    | 真の確率$p(x)$ | 70% | 30% |
    | 推定した確率$q(x)$ | 50% | 50% |
    $$
    D_{KL} (\;p \;|| \;q\;) = 0.7 \log \cfrac{0.7}{0.5} +  0.3 \log \cfrac{0.3}{0.5} = 0.021...
    $$
1. ケースB

    |  | 表の出る確率 | 裏の出る確率 |
    |----|----|----|
    | 真の確率$p(x)$ | 70% | 30% |
    | 推定した確率$q(x)$ | 20% | 80% |
    $$
    D_{KL} (\;p \;|| \;q\;) = 0.7 \log \cfrac{0.7}{0.2} +  0.3 \log \cfrac{0.3}{0.8} = 0.58...
    $$

1. ケースA

    |  | 表の出る確率 | 裏の出る確率 |
    |----|----|----|
    | 真の確率$p(x)$ | 70% | 30% |
    | 推定した確率$q(x)$ | 70% | 30% |
    $$
    D_{KL} (\;p \;|| \;q\;) = 0.7 \log \cfrac{0.7}{0.7} +  0.3 \log \cfrac{0.3}{0.3} = 0
    $$

#### 用途
「**2つの確率分布がどれくらい異なるか**」を表す尺度として使うことができる. 
    

### 5.1.3 KLダイバージェンスと最尤推定の関係
#### 〇 問題設定
「真の確率分布$p_*(x)$」があり、サンプルデータ${x^{(1)}, x^{(2)}, \cdots, x^{(N)}}$を生成したとする. 
#### 〇 目標
パラメータ$\theta$で調整できる確率分布$p_\theta(x)$を使って、$p_*(x)$にできるだけ近い確率分布を作る. 
#### 〇 最尤推定
次の対数尤度を目的関数（＝最大化したい関数）とする. 
$$
\log \prod^N_{n=1} \;p_\theta(x^{(n)}) = \sum^N_{n=1} \;\log p_\theta(x^{(n)}) 
$$
この対数尤度を最大化するパラメータ$\theta$は、次式で表される. 
$$
\hat{\theta} = \argmax_{\theta} \sum^N_{n=1} \;\log p_\theta(x^{(n)})   \quad \text{(推定値には$\;\hat{}\;$ をつける)}  \cdots (5.1)
$$
キーとなるアイデアは、次の言い換えをすること. 
$$
\text{「$p_\theta(x)$を$p_*(x)$にできるだけ近づかせる」}  \xrightarrow[\text{言い換え}]{\ }  \text{「$p_*(x)$と$p_\theta(x)$のKLダイバージェンスを最小にする」}
$$
$p_*(x)$と$p_\theta(x)$のKLダイバージェンスは、次の式で表される.
$$
D_{KL} (\;p_* \;|| \;p_\theta\;) = \int p_*(x) \; \log\cfrac{p_*(x)}{p_\theta(x)} \;dx \quad \text{(真の確率:$p_*(x)$, 推定した確率:$p_\theta(x)$)}  \cdots (5.2)
$$

#### 〇 問題点
$$
\text{(5.2)式を計算したい!!!} \longrightarrow \text{$p_\theta(x)$の具体的な式が不明} \longrightarrow \text{\textbf{モンテカルロ法}を使って近似計算}
$$


#### 〇 モンテカルロ法(Monte Carlo method)
乱数を用いて複雑な確率分布や期待値などの近似値を計算するための手法. ランダムに生成されたサンプルを用いて問題をシミュレートし、それらのサンプルから求めた結果の平均を取ることで、問題の解を近似する. 
##### 手順
1. 求めたい期待値を設定 ($\;p_*(x)\;$は確率密度、$f(x)$は任意の関数).
    $$
    \bm{E}_{p_*(x)}[f(x)] = \int f(x)p_*(x)dx 
    $$
1. 確率分布$\;p_*(x)\;$に基づいてサンプル${x^{(1)}, x^{(2)}, \cdots, x^{(N)}}$を生成.
1. 各データ$\;x^{(i)\;}$における$\;f(x^{(i)})\;$を求め、その平均を計算. 
1. 上記の積分を近似. 
    $$
    \bm{E}_{p_*(x)}[f(x)] = \int f(x)p_*(x)dx \approx \cfrac{1}{N} \sum^N_{n=1} \;f(x^{(n)}) \quad (x^{(n)} \sim p_*(x))
    $$
    - $\;\approx\;$は「近似的に等しい」という記号.
    - $x^{(n)} \sim p_*(x)$は、「$x^{(n)}$が確率分布$p_*(x)$に従っている」ことを示す.

#### KLダイバージェンスへのモンテカルロ法の適用
1. (5.2)式にモンテカルロ法を適用
    $$
    \begin{align*}
    D_{KL} (\;p_* \;|| \;p_\theta\;) &= \int p_*(x) \; \underbrace{\log\cfrac{p_*(x)}{p_\theta(x)}}_{f(x)} \;dx \\ 
    &\underbrace{\approx}_{\text{モンテカルロ法}} \cfrac{1}{N} \sum^N_{n=1} \; \underbrace{\log\cfrac{p_*(x^{(n)})}{p_\theta(x^{(n)})}}_{f(x)} \quad (x^{(n)} \sim p_*(x)) \\
    & =  \cfrac{1}{N} \sum^N_{n=1} \;(\;\underbrace{\log p_*(x^{(n)})}_{\text{$\theta$を含まない項}} - \log p_\theta(x^{(n)}) )
    \end{align*}
    $$
1. 目標は、$ D_{KL} (\;p_* \;|| \;p_\theta\;)$を最小にする$\theta$を求めることなので、$\theta$を含まない項を無視できる. 
    $$
    \begin{align*}
    \argmin_\theta D_{KL} (\;p_* \;|| \;p_\theta\;)  
    &\overset{\text{モンテカルロ法}}{\approx} \argmin_\theta ( -\cfrac{1}{N} \sum^N_{n=1} \; \log p_\theta(x_{n}) ) \\
    &= \argmin_\theta ( - \sum^N_{n=1}\; \log p_\theta(x_{n}) ) \\
    &= \argmax_\theta \sum^N_{n=1} \;\log p_\theta(x_{n}) 
    \end{align*}
    $$
1. 以上より、次式が導かれる. 
    > $$
    \begin{align*}
    \underbrace{\argmin_\theta D_{KL} (\;p_* \;|| \;p_\theta\;)}_{\text{(KLダイバージェンスを最小にする$\theta$)}}  &\overset{\text{モンテカルロ法}}{\approx}  \underbrace{\argmax_\theta \sum^N_{n=1} \;\log p_\theta(x_{n})}_{\text{(対数尤度を最大にする$\theta$)}}
    \end{align*}
    $$

### 5.2 EMアルゴリズムの導出-1

#### 〇 目標
$$
\text{「潜在変数を持つモデル」} \longrightarrow \text{「混合ガウスモデル(GMM)」} \quad \text{という手順で、\textbf{EMアルゴリズム}を適用}
$$

### 5.2.1 潜在変数を持つモデル
#### 〇 一つのデータに関する対数尤度
観測できる確率変数を$\;x\;$, 潜在変数(離散型)を$\;z\;$, パラメータを$\;\theta \;$で表す. このとき、一つのデータに対する対数尤度は、確率の周辺化(p.78)により、次式で表す. 
$$
\log\;p_\theta(x) = \log\;\sum_x p_\theta(x,z)  \quad \text{($\;z\;$が連続型の場合は、$\sum_x \rightarrow \int dz$)}
$$
サンプル$\; \mathcal{D} = \{x^{(1)}, x^{(2)}, \cdots, x^{(N)}\}\;$が得られた場合、対数尤度は次式で表される. 
$$
\begin{align*}
\log\;p_\theta(x) &= \log \big( \;p_\theta(x^{(1)})p_\theta(x^{(2)}) \cdots p_\theta(x^{(N)}) \;\big) \quad \text{($\;\because$ 各データが$p_\theta(x)$に従って、\underline{独立に}生成されると仮定(p.30)\;)}\\
&= \sum^N_{n=1} \log p_\theta(x^{(n)}) \\ 
&= \sum^N_{n=1} \log \sum_{z^{(n)}}p_\theta(x^{(n)}, z^{(n)}) \underset{\text{最大化したい}}{\longrightarrow}\text{「log-sum」の形となり、解析的に解けない...}  \underset{\textbf{EMアルゴリズム}}{\longrightarrow} \text{「sum-log」の形に変換}
\end{align*}
$$

#### 〇 問題点
問題は、対数尤度が次式であること.
$$
\log p_\theta(x) = \log \sum_{z} p_\theta(x,z) \quad \text{が「log-sum」の形になっている}
$$
まずは、確率の乗法定理(p.79)を使って表す. 
$$
\log p_\theta(x) = \log \cfrac{p_\theta(x,z)}{p_\theta(z\;|\;x)} \quad \text{$\big(\; \because p_\theta(z,x) = p_\theta(z\;|\;x)\;p_\theta(x)\;\big)$}
$$
一見、「log-sum」を解決しているように見えるが、条件付確率の$p_\theta(z\;|\;x)$が"厄介者"である. なぜなら、$p_\theta(z\;|\;x)$はベイズの定理より次の式で表されるから。
$$
p_\theta(z\;|\;x) = \cfrac{p_\theta(x,z)}{\sum_z p_\theta(x,z)} \quad \longleftarrow \text{結局、式の分母に「$\sum$(sum)」が登場...}
$$

### 5.2.2 任意の確率分布$\;q(z)\;$
#### 〇 $\;q(z)\;$の導入
以上の問題に対処するため、$\;q(z)\;$を登場させる. $\;q(z)\;$は「任意の確率分布」で、「どのような確率分布であってもよい」ことを指す.

$\log p_\theta(x)$を次のように展開.
$$
\begin{align*}
\log p_\theta(x) 
&= \log \cfrac{p_\theta(x,z)}{p_\theta(z\;|\;x)} \\ 
&= \log \cfrac{p_\theta(x,z)}{p_\theta(z\;|\;x)} \; \cfrac{q(z)}{q(z)} \\
&= \log \cfrac{p_\theta(x,z)}{q(z)} + \log \underbrace{\cfrac{q(z)}{p_\theta(z\;|\;x)}}_{\text{"厄介者"}}\quad\cdots (5.4)
\end{align*}
$$

#### 〇 KLダイバージェンス
KLダイバージェンスがもつ、「2つの確率分布が同じときに最小値の0をとる」という扱いやすい性質を用いる. 

次の目標は、第二項をKLダイバージェンスで表すこと. 
$$
\begin{align*}
&\log p_\theta(x) \\
&= \log p_\theta(x) \sum_z q(z)\quad \text{$\Big($ $\sum_z q(z)=1$をかける $\Big)$}\\
&= \sum_z q(z) \log p_\theta(x) \\
&= \sum_z q(z)\Bigg( \log \cfrac{p_\theta(x,z)}{q(z)} + \log \cfrac{q(z)}{p_\theta(z\;|\;x)} \Bigg) \quad \text{(式(5.4)を代入)} \\ 
&= \sum_z q(z)\log \cfrac{p_\theta(x,z)}{q(z)} + \underbrace{\sum_z q(z)\log \cfrac{q(z)}{p_\theta(z\;|\;x)}}_{\text{KLダイバージェンス}}\\ 
&= \sum_z q(z)\log \cfrac{p_\theta(x,z)}{q(z)} + D_{KL} (\;q(z) \;|| \;p_\theta(z\;|\;x)\;)  \longrightarrow \text{EMアルゴリズムを導くための式}
\end{align*}
$$
---


### 5.3 EMアルゴリズムの導出-2
ここまでで得た式; 
$$
\log p_\theta(x) = \sum_z q(z)\log \cfrac{p_\theta(x,z)}{q(z)} + D_{KL} (\;q(z) \;|| \;p_\theta(z\;|\;x)\;) \quad \cdots (5.5)
$$

### 5.3.1 ELBO(エビデンスの下界)
どのような確率分布であっても、KLダイバージェンスは常に0以上になる(*Jensen*の不等式より)ことより、次式を得る. 
$$
\begin{align*}
\log p_\theta(x) &= \sum_z q(z)\log \cfrac{p_\theta(x,z)}{q(z)} + \underbrace{\sum_z q(z)\log \cfrac{q(z)}{p_\theta(z\;|\;x)}}_{\text{$\geq 0$}}\\
&\geq \sum_z q(z)\log \cfrac{p_\theta(x,z)}{q(z)} \\
&= \text{ELBO}
\end{align*}
$$

ELBOとは、「Evidence Lower BOund」の略で、「エビデンスの下界」と訳される. 
$$
\text{ELBO}(x;q,\theta) =  \underbrace{\sum_z q(z) \log \cfrac{p_\theta(x,z)}{q(z)}}_{\text{「sum-log」の形}}\quad (5.6)
$$
ELBOには、次に示す重要な特徴がある.
- 対数尤度$\;\log p_\theta(x)\;$は常にELBO $(x;q,\theta)$ 以上の値になる
- 式(5.5)のELBO $(x;q,\theta)$ は、「sum-log」の形になっており、解析しやすい. 

従って、ELBO $(x;q,\theta)$を大きくするようにパラメータを更新すれば、 $\log p_\theta(x)$ はそれ以上の値になる. 
$$
\log p_\theta(x) \underset{\text{最適化の対象}}{\longrightarrow} \text{ELBO}(x;q,\theta)
$$

### 5.3.2 EMアルゴリズムへ
ELBOには、2つのパラメータ$\;q(z)\;$(=任意の確率分布)と$\;\theta\;$がある. 
#### 〇 目標
> $\;q(z)\;$と$\;\theta\;$ の最適化( = ELBOが最大となる2つのパラメータを見つける)

#### 〇 手順
同時に最適化することは困難. 従って、片方を固定して、もう片方のパラメータを最適化するという作業を繰り返す. 具体的には、
1. $\;\theta\;$を固定して、$\;q(z)\;$を更新.
1. $\;q(z)\;$を固定して、$\;\theta\;$を更新. 
1. 以上を繰り返す. 

#### 〇 E-step ($\;q(z)\;$の最適化 )
まず、$\;\theta\;$を現状の値で固定( $\;\theta\; = \;\theta_{old}\;$ )として、$\;q(z)\;$の更新を考える. 
- $\;q(z)\;$がどのような確率分布であっても、$\;\log p_\theta(x) \geq \text{ELBO}(x;q,\theta)\;$が成立. 
- $\;\text{ELBO}(x;q,\theta)\;$をなるべく$\;\log p_\theta(x)\;$に近づけたい. 
- 次式をヒントにする.
    $$
    \underbrace{\log p_\theta(x)}_{\text{q(z)に依らない}
    } = \text{ELBO}(x;q,\theta) + \sum_z q(z)\log \cfrac{q(z)}{p_\theta(z\;|\;x)}\quad\longrightarrow \textbf{$q(z)$に依らず、「ELBO項 + KL項 は 一定」}
    $$

- 従って、KL項を小さくするほどELBO項は大きくなる. KL項は、$q(z) = p_\theta(z\;|\;x)$となるときに0となる. 
    $$
    q(z) = p_\theta(z\;|\;x) \iff \text{KL項}=0 \iff \text{ELBO項}\rightarrow\text{最大}
    $$
**更新式**
> $$
    q(z) = p_{\theta_{old}}(z\;|\;x) 
    $$
- $\;\theta\; = \;\theta_{old}\;$ において、ELBOと対数尤度が一致する. 
- E-stepの由来:「Expectation Value(期待値)」として表されるから.
    $$
    \begin{align*}
    \text{ELBO}(x;q=p_{\theta_{old}}(z\;|\;x),\theta)
    &= \sum_z p_{\theta_{old}(z\;|\;x)} \log \cfrac{p_{\theta}(x,z)}{p_{\theta_{old}}(z\;|\;x)}  \quad \big(\because (5.6) \;\big)\\
    &= \bm{E}_{p_{\theta_{old}}(z\;|\;x)}\Bigg[ \;\log \cfrac{p_{\theta}(x,z)}{p_{\theta_{old}}(z\;|\;x)}  \;\Bigg] \quad \cdots (5.8)
    \end{align*} 
    $$

#### 〇 M-step ($\;\theta\;$の最適化 )
ELBOが「sum-log」の形をしているので、解析的に求めることができる.
- (5.8)式の$\;\theta\;$に関する微分を求めて、それを0として方程式を解く.
- M-stepの由来: 「Maximazation(最大化)」の頭文字. ELBO を $\;\theta\;$に関して「最大化」することを意味. 


#### 〇 EMアルゴリズム
1. E-step : $\;q(z)\;$を更新 $\rightarrow$ ELBO を $\log p(x;\theta) $に近づける(イコールにする).
1. M-step : $\;\theta\;$を更新 $\rightarrow$ ELBO の値を上げ、間接的に $\log p(x;\theta) $ の値を押し上げる. 
1. ELBOと $\log p(x;\theta) $ が遠ざかるので、1.のステップを行う. 

#### 〇 EMアルゴリズムにおける対数尤度の単調増加性
- EMアルゴリズムの一回の更新により、対数尤度は常に「単調に増加」する. 
    $$
    \log p(x; \theta_{new}) \geq \log p(x; \theta_{old})
    $$
- EMアルゴリズムの更新を続け、対数尤度が変化しなくなるところで更新を止める(実用上は、閾値を設定). 


### 5.3.3 複数データへの拡張
- N個の観測データ${x^{(1)}, x^{(2)}, \cdots, x^{(N)}}$に対して、EMアルゴリズムを拡張. 
- 各データに対応した「任意の確率分布」を${q^{(1)}, q^{(2)}, \cdots, q^{(N)}}$としてN個用意. 
- 対数尤度とELBOの関係は次の通り ( (5.6)式に対して、「$\sum^N_{n=1}$」を追加 ) . 
    $$
    \begin{align*}
    \sum^N_{n=1} \;\log \sum_{z^{(n)}} p_\theta (x^{(n)}, z^{(n)} ) 
    &\geq \sum^N_{n=1} \text{ELBO} (x^{(n)}; q^{(n)}, \theta) \\ 
    &= \sum^N_{n=1} \sum_{z^{(n)}} q^{(n)} \big( z^{(n)} \big) \log \cfrac{p_\theta (x^{(n)}, z^{(n)})}{ q^{(n)} \big( z^{(n)} \big)}
    \end{align*}
    $$
- N個のデータの場合の更新式. 
    $$
    \begin{align*}
    データが一つのとき : \;&最適な確率分布 \;q(z)\; は \;p_\theta(z\;|\;x).    \\ 
    データが{x^{(1)}, x^{(2)}, \cdots, x^{(N)}}のとき : \;&最適な確率分布 \;q^{(n)}(z)\; は \;p_\theta(z\;|\;x^{(n)})   
    \end{align*}
    $$
- N個のデータ終了判定は、「対数尤度の平均」を用いる. 
    $$
    \cfrac{1}{N} \sum^N_{n=1} \log p(x^{(n)}; \theta)
    $$

#### EMアルゴリズムのまとめ
1. **E-step** : ${q^{(1)}, q^{(2)}, \cdots, q^{(N)}}$ の更新 ($\;\theta\;$は固定). 
1. **M-step** : $\;\theta\;$の更新 (${q^{(1)}, q^{(2)}, \cdots, q^{(N)}}$ は固定). 
1. **終了判定** : 対数尤度（の平均）を計算し、前回の対数尤度と比較する. 


### 5.3.4 $\log p(x; \theta_{new}) \geq \log p(x; \theta_{old})$の証明
- Eステップにおいて、$\;q_{old}(z)=p_{\theta_{old}}(z\;|\;x)\;$で更新したとき、対数尤度とELBOのが一致するので次式が成り立つ. 
    $$
    \log p(x; \theta_{old}) = \text{ELBO}(x;q_{old}, \theta_{old}) \quad (\because (5.5))
    $$
- Mステップでは、右辺を最大化することで$\theta_{new}$を得るので、次式が成り立つ. 
    $$
    \text{ELBO}(x;q_{old}, \theta_{new}) \geq \text{ELBO}(x;q_{old}, \theta_{old})
    $$
- さらに、その次のEステップでは、$q_{new}(z) = p_{\theta_{new}}(z\;|\;x)$で更新するので、次式が成り立つ. 
    $$
    \begin{align*}
    \log p(x; \theta_{new}) 
    &= \text{ELBO}(x;q_{new}, \theta_{new}) \\ 
    &\geq \text{ELBO}(x;q_{old}, \theta_{new}) \\
    &\geq \text{ELBO}(x;q_{old}, \theta_{old}) \\
    &= \log p(x; \theta_{old}) 
    \end{align*}
    $$

### 5.4 GMMとEMアルゴリズム
ここでは, GMMに対してEMアルゴリズムを適用する. 
### 5.4.1 EMアルゴリズムのEステップ
#### 〇 復習
- 観測変数$\;\bm{x}\;$, 潜在変数を$\;z\;$として、$\;z\;$は 1 から K までの離散値を取る確率変数とする. 
- GMMのパラメータ$\;\theta= \{\bm{\phi}, \bm{\mu}, \bm{\Sigma}\}\;$
- これらのパラメータは次のように、$K$個の要素を持つ ( $K$個のカテゴリカル分布確率を持つ ). 
    $$
    \begin{align*}
    \bm{\phi} = \{ \phi_1, \phi_2, \cdots, \phi_K \} \\
    \bm{\mu} = \{ \mu_1, \mu_2, \cdots, \mu_K \} \\
    \bm{\Sigma} = \{ \Sigma_1, \Sigma_2, \cdots, \Sigma_K \} 
    \end{align*}
    $$
- このとき、尤度は次式で表される. 
    $$
    \begin{align*}
    p(\bm{x}; \bm{\theta}) 
    &= \sum^K_{j=1} p(\bm{x}, z=j; \bm{\theta})\\
    &= \sum^K_{j=1} \underbrace{p(z=j; \bm{\theta})}_{\phi_j} \; \underbrace{p(\bm{x}\;|\; z=j; \bm{\theta})}_{\mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j})} \quad  (\because \text{乗法定理})\\
    &= \sum^K_{j=1} \phi_j \;\mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j}) \quad \cdots(*)
    \end{align*}
    $$
- いま、$N$個の観測データ${x^{(1)}, x^{(2)}, \cdots, x^{(N)}}$が得られたとするとき、各データごとの対数尤度の平均は次式で表される. 終了判定はこの値により行う. 
    $$
    \cfrac{1}{N} \sum^N_{n=1} \log p(\bm{x^{(n)}}; \bm{\theta}) = \cfrac{1}{N} \sum^N_{n=1} \log \sum^K_{j=1} \phi_j \;\mathcal{N}(\bm{x^{(n)}}; \bm{\mu_j}, \bm{\Sigma_j}) \quad (\;\because (*)式\;)
    $$

#### 5.4.1 EMアルゴリズムのE-step
- $\{\bm{\phi}, \bm{\mu}, \bm{\Sigma}\}$を固定して、確率分布$\;q^{(n)}(z)\;$を更新する. 
- $\;q^{(n)}(z)\;$は、$\bm{x^{(n)}}$が与えられた時の「条件付き確率」として求められる. 
    $$
    \begin{align*}
    q^{(n)}(z=k) 
    &= p(z=k\;|\;\bm{x^{(n)}}; \bm{\theta}) \quad \text{(条件付き確率)}\\
    &= \cfrac{p(\bm{x^{(n)}}, z=k\;|\; \bm{\theta})}{p(\bm{x^{(n)}}; \theta)}   \text{(確率の乗法定理)} \\ 
    &= \cfrac{\phi_k\;\mathcal{N}(\bm{x}; \bm{\mu_k}, \bm{\Sigma_k})}{\sum^K_{j=1}\phi_j\;\mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j})} \quad(\because (*)式\;)
    \end{align*} 
    $$
- 以降、簡単のために$q^{(n)}(z=k) = q^{(n)}(k)$で表すことにする. 

#### 5.4.2 EMアルゴリズムのM-step
- $\;q^{(n)}\;$を固定して、$\;\bm{\theta}=\{\bm{\phi}, \bm{\mu}, \bm{\Sigma}\}\;$を更新( 具体的には、ELBOを最大化する$\{\bm{\phi}, \bm{\mu}, \bm{\Sigma}\}$を求める ). 
- ELBOは次式で表される. 
    $$
    \begin{align*}
    &\sum^N_{n=1} \text{ELBO}(\bm{x^{(n)}}; q^{(n)}, \bm{\theta})    \\
    &= \sum^N_{n=1} \sum^K_{j=1}\;q^{(n)}(j)\log \cfrac{p(\bm{x^{(n)}}, z^{(n)}=j\;|\; \bm{\theta})}{q^{(n)}(j)} \\
    &= \sum^N_{n=1} \sum^K_{j=1}\;q^{(n)}(j)\log \cfrac{\phi_j\;\mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j})}{q^{(n)}(j)} \quad (\because (*)式\;)\\
    &= \underbrace{\sum^N_{n=1} \sum^N_{j=1}\;q^{(n)}(j)\log \phi_j\;\mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j})}_{これを最大化したい} - \underbrace{\sum^N_{n=1} \sum^N_{j=1}\;q^{(n)}(j)\log q^{(n)}(j)}_{\{\bm{\phi}, \bm{\mu}, \bm{\Sigma}\}とは無関係}
    \end{align*}
    $$
- 上式で、第2項には$\{\bm{\phi}, \bm{\mu}, \bm{\Sigma}\}$に依らないため、目的関数(=最大化する対象)$J$は第1項となる. 
    $$
    \begin{align*}
    J(\bm{\phi}, \bm{\mu}, \bm{\Sigma}) 
    &= \sum^N_{n=1} \sum^K_{j=1}\;q^{(n)}(j)\log \phi_j\;\mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j}) \\ 
    &= \sum^N_{n=1} \sum^K_{j=1}\;q^{(n)}(j) \;\Big( \log \phi_j + \log \mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j}) \Big)    \quad \cdots(5.9)
    \end{align*}
    $$
- $J(\bm{\phi}, \bm{\mu}, \bm{\Sigma}) $ が最大となるパラメータを求める. 
    1. $\bm{\mu}$の更新
        - $\{{\bm{\mu_1}, \bm{\mu_2}, \cdots, \bm{\mu_k}, \cdots, \bm{\mu_K}}\}$により構成. k番目の$\bm{\mu_k}$に関しての微分（勾配）を求める. 
            $$
            \begin{align*}
            \cfrac{\partial J}{\partial \bm{\mu_k}}
            &= \cfrac{\partial}{\partial \bm{\mu_k}} \sum^N_{n=1} \sum^K_{j=1}\;q^{(n)}(j) \;\Big( \log \phi_j + \log \mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j}) \Big) \\
            &= \cfrac{\partial}{\partial \bm{\mu_k}} \sum^N_{n=1} \sum^K_{j=1}\;q^{(n)}(j) \;\log \mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j}) \\ 
            &= \sum^N_{n=1}\;q^{(n)}(k) \cfrac{\partial}{\partial \bm{\mu_k}}\;\log \mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j}) \quad \cdots (5.10)
            \end{align*}
            $$
        - $\bm{\mu_k}$に関しての微分は、$\bm{\mu_k}$が含まれない項はゼロになる. 
            1. $\;j=1$から$K$に対して、$\cfrac{\partial}{\partial \bm{\mu_k}} \log\phi_j = 0$
            1. $\;j\neq k$ の$\;j=1\;$から$\;K\;$に対して、$\cfrac{\partial}{\partial \bm{\mu_k}} \log \mathcal{N}(\bm{x^{(n)}}; \bm{\mu_j}, \bm{\Sigma_j}) = 0$
        - (5.10)式に出てくる次の項について、(A.6)を利用. 
            $$
            \cfrac{\partial}{\partial \bm{\mu_k}} \log \mathcal{N}(\bm{x^{(n)}}; \bm{\mu_j}, \bm{\Sigma_j}) = \bm{\Sigma^{-1}}(\bm{x}-\bm{\mu})
            $$
        - よって、
            $$
            \cfrac{\partial J}{\partial \bm{\mu_k}} = \sum^N_{n=1} q^{(n)}(k) \;\bm{\Sigma^{-1}_k}(\bm{x^{(n)}}-\bm{\mu_k})
            $$
        - $\cfrac{\partial J}{\partial \bm{\mu_k}}=\bm{0}$とすることで、
            $$
            \begin{align*}
            &\sum^N_{n=1} q^{(n)}(k) \;\bm{\Sigma^{-1}_k}(\bm{x^{(n)}}-\bm{\mu_k}) \\
            &\iff \textcolor{OrangeRed}{\Sigma_k} \sum^N_{n=1} q^{(n)}(k) \;\bm{\Sigma^{-1}_k}(\bm{x^{(n)}}-\bm{\mu_k}) = \textcolor{OrangeRed}{\Sigma_k} \bm{0}    \quad \text{(両辺に左から$\;\textcolor{OrangeRed}{\Sigma_k}\;$をかける)} \\
            &\iff \sum^N_{n=1} q^{(n)}(k) \;\underbrace{\textcolor{OrangeRed}{\Sigma_k} \;\bm{\Sigma^{-1}_k}}_{=I}\;(\bm{x^{(n)}}-\bm{\mu_k}) =\bm{0} \quad \text{($\;q^{(n)}(k)\;$はスカラ)} \\
            &\iff \sum^N_{n=1} q^{(n)}(k) \;(\bm{x^{(n)}}-\bm{\mu_k}) =\bm{0} \\ 
            &\iff \sum^N_{n=1} q^{(n)}(k) \;\bm{\mu_k} = \sum^N_{n=1} q^{(n)}(k) \;\bm{x^{(n)}} \\ 
            &\iff \bm{\mu_k} \Big( \sum^N_{n=1} q^{(n)}(k)  \Big) = \sum^N_{n=1} q^{(n)}(k) \;\bm{x^{(n)}}  \quad (\text{$\;\bm{\mu_k}\;$は$\;\sum^N_{n=1}\;$}とは独立) \\
            &\therefore \bm{\mu_k} = \cfrac{ \sum^N_{n=1} q^{(n)}(k)\;\bm{x^{(n)}} }{ \sum^N_{n=1} q^{(n)}(k) } \quad \cdots(5.11)
            \end{align*}
            $$
    1. 共分散行列$\bm{\Sigma}$の更新
        - (A.15)式を用いる
            $$
            \bm{\Sigma_k} = \cfrac{\sum^N_{n=1} q^{(n)}(k)\;(\bm{x^{(n)}}-\bm{\mu_k})\;(\bm{x^{(n)}}-\bm{\mu_k})^T}{\sum^N_{n=1} q^{(n)}(k)} \quad \cdots(5.12)
            $$
    1. $\bm{\phi}=\{\phi_1, \phi_2, \cdots, \phi_K\}$の更新
        - 目的関数である式(5.9)において、$\bm{\phi}$に関する項は次式で表される. 
            $$
            \sum^N_{n=1} \sum^K_{j=1} q^{(n)}(j)\log \phi_j 
            $$
        - ただし、$\phi$の最適化は$\bm{\mu}$のときと違い、次の制約が存在. 
            $$
            \sum^K_{j=1} \phi_j = 1 \longrightarrow 「ラグランジュの未定乗数法」を用いて解く. 
            $$
        - いま、変数は$\phi_k$であり、ラグラジアン$L$は、次式で表される. 
            $$
            L = \sum^N_{n=1} \sum^K_{j=1} q^{(n)}(j)\;\log \phi_j + \beta  \underbrace{ \Big(\sum^K_{j=1} \phi_j -1 \Big)}_{g(x)=0にあたる「制約条件」} 
            $$
        - $\partial L / \partial \beta = 0$にあたる式. 
            $$
            \sum^K_{j=1} \phi_j -1 = 0 \iff \sum^K_{j=1} \phi_j = 1 \quad \text{(制約条件そのもの)} \quad \cdots (**)
            $$
        - $\partial L / \partial x = 0$にあたる式. 
            $$
            \begin{align*}
            \cfrac{\partial L}{\partial \phi_k} 
            &= \sum^N_{n=1} q^{(n)}(k) \cfrac{\partial}{\partial \phi_k} \log \phi_k + \beta \\
            &= \sum^N_{n=1} \cfrac{q^{(n)}(k)}{\phi_k} + \beta \\
            &= 0 \\ 
            &\iff \phi_k = \cfrac{\sum^N_{n=1} q^{(n)}(k) }{ -\beta }   \quad (5.13)
            \end{align*}
            $$
        - (5.13)式と(**)を共に満たすならば( (5.13)式に(\*\*)式を代入すればよく ),
            $$
            \begin{align*}
            &\sum^K_{j=1} \cfrac{\sum^N_{n=1} q^{(n)}(j) }{ -\beta } = 1 \\ 
            &\iff -\beta = \sum^K_{j=1} \sum^N_{n=1} q^{(n)}(j) \\ 
            &\iff -\beta = \sum^N_{n=1} \underbrace{\sum^K_{j=1} q^{(n)}(j)}_{1} \quad \cdots(5.14)\\ 
            &\iff -\beta = N   \quad \Big(\because q^{(n)}(z)はzの確率分布であるから、zに関する確率の和は\;\sum^K_{j=1}q^{(n)}(z=j)=1\;となる \Big)
            \end{align*}
            $$
        - (5.13)式に$\;-\beta = N\;$を代入する. 
            $$
            \phi_k = \cfrac{1}{N}\sum^N_{n=1} q^{(n)}(k)
            $$


#### 更新式と終了判定
以上、GMMに対するEMアルゴリズムの導出をまとめると、つぎのようになる. 
1. **E-step**

    各$n,k$に対して次の値を計算. 
    $$
    q^{(n)}(k) = \cfrac{\phi_k\;\mathcal{N}(\bm{x^{(n)}}; \bm{\mu_k}, \bm{\Sigma_k})}{\sum^K_{j=1}\phi_j\;\mathcal{N}(\bm{x^{(n)}}; \bm{\mu_j}, \bm{\Sigma_j})}
    $$
1. **M-step**

    各$k$に対して次の値を計算. 
    $$
    \begin{align*}
    &\phi_k = \cfrac{1}{N}\sum^N_{n=1} q^{(n)}(k) \\ 
    &\bm{\mu_k} = \cfrac{ \sum^N_{n=1} q^{(n)}(k)\;\bm{x^{(n)}} }{ \sum^N_{n=1} q^{(n)}(k) } \\
    &\bm{\Sigma_k} = \cfrac{\sum^N_{n=1} q^{(n)}(k)\;(\bm{x^{(n)}}-\bm{\mu_k})\;(\bm{x^{(n)}}-\bm{\mu_k})^T}{\sum^N_{n=1} q^{(n)}(k)}
    \end{align*}
    $$

1. **終了判定**

    次の対数尤度を計算し、前回の対数尤度と比較する. 
    $$
    \cfrac{1}{N} \sum^N_{n=1} \log \sum^K_{j=1} \phi_j \;\mathcal{N}(\bm{x}; \bm{\mu_j}, \bm{\Sigma_j})
    $$

```python
import os 
import numpy as np
import matplotlib.pyplot as plt

__file__ = ''

path = os.path.join(os.path.dirname(__file__), 'old_faithful.txt')
xs  = np.loadtxt(path)
print(xs.shape) # (272, 2)

# params
phis = np.array([0.5, 0.5]) 
mus = np.array([[0.0, 50.0], [0.0, 100.0]])
covs = np.array([np.eye(2), np.eye(2)]) # 2つの2x2単位行列

K = len(phis) # 2
N = len(xs) # 272
MAX_ITERS = 100
THRESHOLD = 1e-4     # thresholdのミス？？
```

### 5.5 EMアルゴリズムの実装

```python
# 正規分布 ( 書籍では(3.1)式 )
def multivariate_normal(x, mu, cov):
    det = np.linalg.det(cov)
    inv = np.linalg.inv(cov)
    d = len(x)
    z = 1 / np.sqrt((2 * np.pi) ** d * det)
    y = z * np.exp((x - mu).T @ inv @ (x-mu) / -2.0)
    return y

# 混合ガウスモデル( 書籍では(4.3)式 )
def gmm(x, phis, mus, covs):
    K = len(phis)
    y = 0
    for k in range(K):
        phi, mu, cov = phis[k], mus[k], covs[k]
        y += phi * multivariate_normal(x, mu, cov)
    return y

# 対数尤度を求める関数
def likelihood(xs, phis, mus, covs):
    eps = 1e-8 # log(0)を防ぐための微小値
    L = 0
    N = len(xs)
    for x in xs:
        y = gmm(x, phis, mus, covs)
        L += np.log(y + eps)
    return L / N
```

関数likelihood()は次の数式を表す:
$$
    \cfrac{1}{N} \sum^N_{n=1} \log \sum^K_{j=1} \phi_j \mathcal{N} ( \bm{x}^{(n)};\bm{\mu_j}, \bm{\Sigma_j})
$$
関数内での使用変数の説明:
$$
\begin{align*}
    y &= \sum^K_{j=1} \phi_j \mathcal{N} ( \bm{x}^{(n)};\bm{\mu_j}, \bm{\Sigma_j}) \\
    L &=  \sum^N_{n=1} \log y 
\end{align*}
$$

--- 


### E ステップ
各n, k に対して次の計算をする: 
$$
    q^{(n)}(k) = \cfrac{\phi_k \mathcal{N} (\bm{x^{(n)}}; \bm{\mu_k}, \bm{\Sigma_k})}{\sum^K_{j=1} \phi_j \mathcal{N} (\bm{x^{(n)}}; \bm{\mu_j}, \bm{\Sigma_j})}
$$

### M ステップ
各k に対して次の計算をする: 
$$
\begin{align*}
    \phi_k &= \cfrac{1}{N}\sum^N_{n=1} q^{(n)}(k)   \\
    \bm{\mu_k} &= \cfrac{\sum^N_{n=1} q^{(n)}(k) \bm{x^{(n)}}}{\sum^N_{n=1} q^{(n)}(k)} \\ 
    \bm{\Sigma_k} &= \cfrac{\sum^N_{n=1} q^{(n)}(k) (\bm{x^{(n)}} - \bm{\mu_k})(\bm{x^{(n)}} - \bm{\mu_k})^T}{\sum^N_{n=1} q^{(n)}(k)}
\end{align*}
$$

```python
current_likelihood =  likelihood(xs, phis, mus, covs)

for iter in range(MAX_ITERS):
    # E-step
    qs = np.zeros((N, K))   # qs(複数形)
    for n in range(N):
        x = xs[n]
        for k in range(K):
            phi, mu, cov = phis[k], mus[k], covs[k]
            qs[n, k] = phi * multivariate_normal(x, mu, cov)
        qs[n] /= gmm(x, phis, mus, covs)  # qs[n] = qs[n] / gmm(...) 

    # M-step
    qs_sum = qs.sum(axis=0)     # 先に計算しておく
    for k in range(K):
        # 1. phis
        phis[k] = qs_sum[k] / N

        # 2. mus
        c = 0
        for n in range(N):
            c += qs[n, k] * xs[n]
        mus[k] = c / qs_sum[k]

        # 3. covs
        c = 0
        for n in range(N):
            z = xs[n] - mus[k]
            z = z[:, np.newaxis] # 数式に合わせるため列ベクトルへ(形状を(D,1)にする)
            c += qs[n, k] * z @ z.T
            # print(c.shape)
        covs[k] = c / qs_sum[k]

    # 終了判定
    print(f"{current_likelihood:.3f}") # 対数尤度を出力 (単調増加している)

    next_likelihood = likelihood(xs, phis, mus, covs)
    diff = np.abs(next_likelihood - current_likelihood) # 差分の絶対値
    if diff < THRESHOLD:
        break
    current_likelihood = next_likelihood
```

```python
FontSize = 18
plt.rcParams.update({'font.size': FontSize})

# ------ GMMの可視化 ------
def plot_contour(w, mus, covs): #
    x = np.arange(1, 6, 0.1)
    y = np.arange(40, 100, 1)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            x = np.array([X[i, j], Y[i, j]])
            for k in range(len(mus)):
                mu, cov = mus[k], covs[k]
                Z[i, j] += w[k] * multivariate_normal(x, mu, cov)

    cs = plt.contour(X, Y, Z) # 等高線の情報を取得
    plt.clabel(cs, inline=True, fontsize=FontSize)  # 等高線を表示



plt.figure(figsize = (10, 8))
plt.scatter(xs[:,0], xs[:,1])
plot_contour(phis, mus, covs)
plt.xlabel('Eruptions(Min)')
plt.ylabel('Waiting(Min)')
plt.show()
```

### 何をやってきたか？

生データに対して「**2つの正規分布の組み合わせ**(多峰性分布)」で表現できると仮定し、EMアルゴリズムでそのパラメータ（平均$\mu$・分散$\Sigma$）を推定した。
結果、「データは2つの正規分布に沿うように分かれている」ことが可視化できた. 

- データ
    - `old_faithful.txt` の 噴射継続時間(Eruptions) と 噴出の間隔(Waiting)
- GMMのEMアルゴリズムの実装
    - **$N$** : データ数
    - **$K$** : クラスタ数(=2)
    - **E-step**
        - 各n,kに対して, $q^{(n)}(k)$の更新.
        - N個のデータ点それぞれが、K個のクラスタそれぞれに属する確率を最適化. 
    - **M-step**
        - 各kに対して, $\bm{\theta} = \{ \bm{\phi}, \bm{\mu}, \bm{\Sigma} \}$の更新.
        - K個の正規分布それぞれのパラメータを最適化. 
- 結果について
    - 散布図 : データ点が二つのクラスタに分かれている
        - 左下 : 短い噴出 + 短い間隔
        - 右上 : 長い噴出 + 長い間隔
    - 分布
        - データが、二つの正規分布にフィットしている. 
        

```python
# 3D表示
def plot_surface(w, mus, covs):
    x = np.arange(1, 6, 0.1)
    y = np.arange(40, 100, 1)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            p = np.array([X[i, j], Y[i, j]])
            for k in range(len(mus)):
                mu, cov = mus[k], covs[k]
                Z[i, j] += w[k] * multivariate_normal(p, mu, cov)

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(xs[:,0], xs[:,1], zs=0, zdir='z', c='k', alpha=0.5)
    ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)
    ax.set_xlabel('Eruptions(Min)')
    ax.set_ylabel('Waiting(Min)')
    ax.set_zlabel('Density')
    plt.show()

plot_surface(phis, mus, covs)
```

### 5.5.3 データ生成

```python
# 計算したmu, covに沿って(正規分布から)データを生成
N = 500
new_xs = np.zeros((N,2))
for n in range(N): 
    k = np.random.choice(2, p=phis) # phiの確率分布に基づいて 0 または 1 のいずれかをランダムに選択(= クラスタ)
    mu, cov = mus[k], covs[k]
    new_xs[n] = np.random.multivariate_normal(mu, cov)  # 対応するガウス分布からデータを生成
```

## 付録B 
### B.1 凸関数とイェンセンの不等式
#### 〇関数の内分点
- 「$f(x)$が凸関数である」$\iff$ 「$f(x)$のすべての弦が関数より上、もしくは関数上にある」
    - 弦 : 関数上の任意の二点間の線分
    - 数式

        任意の二つの点$x_1, x_2$に対して、$\phi$が$[0,1]$の間で以下の上限が成り立つとき凸関数という. 
        
        $
        \phi f(x_1) + (1-\phi)f(x_2) \geq f(\phi x_1 + (1-\phi)x_2) \quad (B.1)
        $

![](../function_and_string.png)


- 式(B.1)の$\; \phi x_1 + (1-\phi)x_2 \;$は、点$x_1$と$x_2$の間を$\; 1-\phi:\phi \;$の比率で内分する点を表す.
- 同様に、$\phi f(x_1) + (1-\phi)f(x_2)$ は、点$f(x_1)$と$f(x_2)$の間を$\; 1-\phi:\phi \;$の比率で内分する点を表す. 

![](../naibunten.png)


- これらより、(B.1)式は次のように可視化できる. 
![](../kashika.png)

- 「弦上の点が関数上の点よりも上側、もしくは関数上にある」が、任意の二点に対して成り立つ時、$f(x)$は凸関数であるという. 
- 凸関数であるかどうかは、関数の二階微分をもとめることでもわかる. 
    - すべての$x$でゼロ以上であれば、凸関数である. 

#### 〇 イェンセンの不等式（複数の内分点への拡張）
- 次のとき、イェンセンの不等式が成り立つ. 
    1. $\{ \phi_1, \phi_2, \cdots, \phi_N \}$という$N$個の正の実数があり、$\sum^N_{n=1}=1$を満たす.
    1. $f(x)$が凸関数である. 
    $$
    \sum^N_{n=1} \underbrace{\textcolor{Salmon}{\phi_n}}_{重み} f(x_n) \geq f\Big( \sum^N_{n=1}\underbrace{\textcolor{Salmon}{\phi_n}}_{重み} x_n \Big)
    $$
    左辺は、各$\textcolor{RoyalBlue}{f}(x_n)$に重み$\textcolor{Salmon}{\phi_n}$がかかり、和が求まる. 右辺は$x_n$に重み$\textcolor{Salmon}{\phi_n}$がかかり、その和を求めた後で関数$\;\textcolor{RoyalBlue}{f}\;$が適用される.
    $$
    \begin{align*}
    \sum^N_{n=1} \textcolor{Salmon}{\phi_n} \textcolor{RoyalBlue}{f}(x_n) = \textcolor{Salmon}{\phi_1} \textcolor{RoyalBlue}{f}(x_1) + \textcolor{Salmon}{\phi_2} \textcolor{RoyalBlue}{f}(x_2) + \cdots + \textcolor{Salmon}{\phi_N} \textcolor{RoyalBlue}{f}(x_N) \\ 
    \textcolor{RoyalBlue}{f}\Big( \sum^N_{n=1}\textcolor{Salmon}{\phi_n} x_n \Big) = \textcolor{RoyalBlue}{f}(\textcolor{Salmon}{\phi_1} x_1 + \textcolor{Salmon}{\phi_1} x_1 + \cdots + \textcolor{Salmon}{\phi_N} x_N)
    \end{align*}
    $$

### B.2 凹関数とlog関数
- 凹関数
    - 凸関数とは逆の性質. 
    - すべての弦が関数よりも下もしくは関数上にある. 
- 関数$f(x)$が凸関数ならば、それにマイナスをつけた$-f(x)$は凹関数. 
- 逆に、関数$f(x)$が凹関数ならば、それにマイナスをつけた$-f(x)$は凸関数. 

#### $y=\log x$が凹関数であることの証明
- $f(x)=-log x$の一階微分、二階微分を求める.
    $$
    \begin{align*}
    f'(x) = -\cfrac{1}{x} \\ 
    f''(x) = \cfrac{1}{x^2}
    \end{align*}
    $$
- 二階微分$f''(x) > 0$となるので、$-\log x$は凸関数 $\Rightarrow$ $\log x$は凹関数. 

#### 凹関数の場合でも、イェンセンの不等式は成り立つか？
- $f(x) = -\log x$を考える. 
- $-\log x$は凸関数であるから、イェンセンの不等式(B.2)が成り立つ. 
    $$
    \begin{align*}
    &\sum^N_{n=1} -\phi_n \log x_n \geq -\log\sum^N_{n=1}\phi_n x_n \\ 
    \iff&\sum^N_{n=1} \phi_n \log x_n \leq \log\sum^N_{n=1}\phi_n x_n
    \end{align*}
    $$
- 即ち、凹関数では(B.2)の不等号が逆になるだけ.

### B.3 ELBOの導出
イェンセンの不等式を用いると、ELBOを導くことができる. 
- $q(z)$は確率分布であり、$0\geq q(z)\geq 1$かつ$\sum_z q(z)=1$を満たすとすれば、任意の関数$f(x)$において、イェンセンの不等式が成立. 
    $$
    \sum_z q(z)\log f(x) \;\leq \;\log \sum_z q(z) f(z)
    $$
- 対数尤度が$\log p(x;\theta)$で表されるとき、イェンセンの不等式により次のように展開できる. 
    $$
    \begin{align*}
    \log p(x;\theta) 
    &= \log \sum_z p(x,z;\theta) \\ 
    &= \log \sum_z p(x,z;\theta) \cfrac{q(z)}{q(z)} \\
    &= \log \sum_z q(z) \cfrac{p(x,z;\theta)}{q(z)} \\
    &\geq \underbrace{\sum_z q(z) \log \cfrac{p(x,z;\theta)}{q(z)}}_{ELBO} \quad \text{(イェンセンの不等式)}

    \end{align*}
    $$
